## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_grouped_purchase_df = purchase_df.groupby(["Gender"])
gender_grouped_purchase_df["Purchase ID"].count().head(10)
#gender_grouped_purchase_df.head()
#Total putchases by gender
total_purchase_value = gender_grouped_purchase_df["Price"].sum()
total_purchase_value.head()
re_total_purchase_value = total_purchase_value.map("${:,.2f}".format)
re_total_purchase_value.head()

#Average purchases by gender
avg_purchase_price = gender_grouped_purchase_df["Price"].mean()
avg_purchase_price.head()
re_avg_purchase_price = avg_purchase_price.map("${:,.2f}".format)
re_avg_purchase_price.head()

normal_total = total_purchase_value/gender_grouped_purchase_df["Purchase ID"].count()
re_normal_total = normal_total.map("${:,.2f}".format)
re_normal_total.head()

clean_gender_purchase_data_df =pd.DataFrame(gender_grouped_purchase_df["Purchase ID"].count())
clean_gender_purchase_data_df["Average Purchase Price"] =re_avg_purchase_price
clean_gender_purchase_data_df["Total Purchase Value"] = re_total_purchase_value
clean_gender_purchase_data_df["Avg Total Purchase per Person"] = re_normal_total
clean_gender_purchase_data_df

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#bins
age_bins= [0, 9.0, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 9999]
#Bin names
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
group_by_age_purchase_df = purchase_df
group_by_age_purchase_df["Age Summary"] = pd.cut(group_by_age_purchase_df["Age"], age_bins, labels=group_names)
group_by_age_purchase_df
# group from bins
group_by_age_purchase_df = group_by_age_purchase_df.groupby("Age Summary")
group_by_age_purchase_df.count()
#Dataframe
sum_by_age_df =pd.DataFrame(group_by_age_purchase_df.count())
sum_by_age_df

sum_by_age_df["Purchase ID"] =(sum_by_age_df["Purchase ID"]/total_players)*100
sum_by_age_df

sum_by_age_df["Purchase ID"] = sum_by_age_df["Purchase ID"].map("{:,.2f}%".format)
sum_by_age_df

#Formatting
re_sum_by_age_df = sum_by_age_df[["Purchase ID", "SN"]]
re_sum_by_age_df

re_group_by_age_summary_df =re_sum_by_age_df.rename(columns={"Purchase ID": "Percentage of Players", "SN":"Total Count"})
re_group_by_age_summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN = purchase_df.groupby(purchase_df["SN"])
s_name = SN["SN"].unique()
#Purchase count
sname_count = SN['Age'].count()
ave_purchase = round(SN['Price'].mean(),2)
total_purchase_by_sname = SN['Price'].sum()
top_spender = {"SN": s_name, "Purchase Count": sname_count, "Average Purchase Price": ave_purchase, "Total Purchase Value": total_purchase_by_sname}
re_top_spender= pd.DataFrame(top_spender)
re_top_spender = re_top_spender.set_index("SN")
re_top_spender = re_top_spender.sort_values("Total Purchase Value", ascending = False)
re_top_spender = re_top_spender[["Purchase Count", "Average Purchase Price","Total Purchase Value"]]
re_top_spender.iloc[:5]
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item Id
grp_item_id = purchase_df.groupby(purchase_df["Item ID"])
grp_items_id= grp_item_id["Item ID"].unique()
item_name = grp_item_id["Item Name"].unique()
#Purchase count
item_purchase = grp_item_id["Age"].count()
item_price = grp_item_id["Price"].unique()
#Total purchase value
item_total_purchase =grp_item_id["Price"].sum()
item_df = {"Item ID":grp_items_id,"Item Name":item_name, "Item Price":item_price, "Purchase Count":item_purchase,"Total Purchase Value": item_total_purchase}
re_item_df = pd.DataFrame(item_df)
re_item_df = re_item_df.set_index("Item ID")
re_item_df = re_item_df.sort_values("Purchase Count", ascending=False)
re_item_df = re_item_df[["Item Name", "Purchase Count","Item Price","Total Purchase Value"]]
re_item_df.iloc[:5]

```
